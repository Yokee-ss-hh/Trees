```python
class MinHeap:
    def __init__(self,capacity):
        self.heap = [None]*capacity
        self.size = 0
        self.capacity = capacity

    def parent(self,i):
        return (i-1)//2

    def left_child(self,i):
        return (2*i)+1

    def right_child(self,i):
        return (2*i)+2 
    
    def insert(self, ele):
        if self.size == self.capacity:
            raise Exception("Heap is full: Cannot insert")
        else:
            self.size += 1
            i = self.size - 1
            self.heap[i]= ele
            while i > 0 and self.heap[self.parent(i)] > self.heap[i]:
                self.heap[self.parent(i)], self.heap[i] = self.heap[i], self.heap[self.parent(i)]
                i = self.parent(i)
    
    def extract_heapify(self,i):
        l = self.left_child(i)
        r = self.right_child(i)
        smallest = i
        if l < self.size and self.heap[l] < self.heap[i]:
            smallest = l 
        if r < self.size and self.heap[r] < self.heap[smallest]:
            smallest = r 
        if smallest != i:
            self.heap[i], self.heap[smallest] = self.heap[smallest], self.heap[i]
            self.extract_heapify(smallest)
    
    def extract_min(self):
        if self.size <= 0:
            raise Exception("Cannot extract minimum value: Empty heap")
        elif self.size == 1:
            self.size -= 1
            return self.heap[0]
        else:
            to_return = self.heap[0]
            self.heap[0] = self.heap[self.size-1]
            self.size -= 1
            self.extract_heapify(0)
            return to_return

    def delete(self, ele):
        index = self.search(ele)
        self.heap[index] = float("-inf")
        while index > 0 and self.heap[self.parent(index)] > self.heap[index]:
            self.heap[self.parent(index)], self.heap[index] = self.heap[index], self.heap[self.parent(index)]
            index = self.parent(index)
        self.extract_min()

    def print_heap(self):
        for x in range(0,self.size,1):
            print(self.heap[x], end=" ")
        print(" ")

    def search(self, key):
        index = 0
        for i in range(0, self.size):
            if self.heap[i] == key:
                return index 
            else:
                index += 1
        raise Exception("Key not found") 


minheap = MinHeap(6)
minheap.insert(9)
minheap.insert(1)
minheap.insert(8)
minheap.insert(2)
minheap.insert(7)
minheap.insert(3)
minheap.print_heap()
print(minheap.extract_min())
minheap.print_heap()
```
