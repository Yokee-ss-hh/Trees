```python
class Node:
    def __init__(self,data):
        self.data = data
        self.left = None 
        self.right = None 

    #          1
    #        /   \
    #       2     3
    #     /   \  /  \
    #    4    5 6    7
    
class BinaryTree:
    def __init__(self):
        self.root = None 

    def level_order_insertion(self,key):
        if self.root is None:
            self.root = Node(key)
        else:
            queue = []
            queue.append(self.root)
            while len(queue):
                temp = queue.pop(0)
                if not temp.left:
                    temp.left = Node(key)
                    break
                else:
                    queue.append(temp.left)
                if not temp.right:
                    temp.right = Node(key)
                    break 
                else:
                    queue.append(temp.right)

    def level_order_traverse(self):
        if self.root is None:
            raise Exception("No nodes available")
        else:
            queue = []
            queue.append(self.root)
            while queue:
                temp = queue.pop(0)
                print(temp.data)
                if temp.left:
                    queue.append(temp.left)
                if temp.right:
                    queue.append(temp.right)
    
    def level_order_traversal_level_by_level(self):
        if self.root is None:
            raise Exception("No nodes available")
        else:
            queue = []
            queue.append(self.root)
            while queue:
                count = len(queue)
                while count:
                    temp = queue.pop(0)
                    print(temp.data, end=" ")
                    if temp.left:
                        queue.append(temp.left)
                    if temp.right:
                        queue.append(temp.right)
                    count -= 1
                print("")
    
    def pre_order_traversal_recursive(self, dummy):
        if dummy is not None:
            print(dummy.data, end=" ")
            self.pre_order_traversal_recursive(dummy.left)
            self.pre_order_traversal_recursive(dummy.right)
    
        
    def in_order_traversal_recursive(self, dummy):
        if dummy is not None:
            self.in_order_traversal_recursive(dummy.left)
            print(dummy.data, end=" ")
            self.in_order_traversal_recursive(dummy.right)
    

    def post_order_traversal_recursive(self, dummy):
        if dummy is not None:
            self.post_order_traversal_recursive(dummy.left)
            self.post_order_traversal_recursive(dummy.right)
            print(dummy.data, end=" ")
    
    
    def pre_order_traversal_iterative(self):
        if self.root is None:
            return
        else:
            stack = []
            stack.append(self.root)
            while stack:
                node = stack.pop()
                print(node.data, end=" ")
                if node.right:
                    stack.append(node.right)
                if node.left:
                    stack.append(node.left)
    

    def in_order_traversal_iterative(self):
        if self.root is None:
            return 
        else:
            stack = []
            current = self.root
            while current is not None or stack:
                if current is not None:
                    stack.append(current)
                    current = current.left
                else:
                    current = stack.pop()
                    print(current.data,end=" ")
                    current = current.right

    
    def post_order_traversal_iterative_one_stack(self):
        if self.root is None:
            return 
        else:
            current = self.root
            stack = []
            while current is not None or stack:
                if current is not None:
                    stack.append(current)
                    current = current.left 
                else:
                    temp = stack[-1].right
                    if temp == None:
                        temp = stack.pop()
                        print(temp.data, end=" ")
                        while stack and temp == stack[-1].right:
                            temp = stack.pop()
                            print(temp.data, end=" ")
                    else:
                        current = temp 
    

    def post_order_traversal_iterative_two_stacks(self):
        if self.root is None:
            return 
        else:
            current = self.root 
            stack1, stack2 = [], []
            stack1.append(current)
            while stack1:
                node = stack1.pop()
                stack2.append(node)
                if node.left:
                    stack1.append(node.left)
                if node.right:
                    stack1.append(node.right)
        while stack2:
            print(stack2.pop().data, end=" ")

    
    def delete_node(self, key):
        if self.root is None:
            raise Exception("Tree is empty")
        elif self.root.left is None and self.root.right is None:
            if self.root.data == key:
                self.root = None 
            else:
                raise Exception("Key not found!!")
        else:
            current = self.root 
            queue = []
            queue.append(current)
            temp = None 
            last = None 
            key_node = None 
            while len(queue):
                temp = queue.pop(0)
                if temp.data == key:
                    key_node = temp 
                if temp.left:
                    last = temp
                    queue.append(temp.left)
                if temp.right:
                    last = temp
                    queue.append(temp.right)
            if key_node is not None:
                key_node.data = temp.data 
                if last.right == temp:
                    last.right = None 
                else:
                    last.left = None 
            else:
                raise Exception("Key not found!!")


bt = BinaryTree()
bt.level_order_insertion(1)
bt.level_order_insertion(2)
bt.level_order_insertion(3)
bt.level_order_insertion(4)
bt.level_order_insertion(5)
bt.level_order_insertion(6)
bt.level_order_insertion(7)
bt.level_order_traversal_level_by_level()
# bt.level_order_traversal_level_by_level()
# bt.pre_order_traversal_recursive(bt.root)
# bt.in_order_traversal_recursive(bt.root)
# bt.post_order_traversal_recursive(bt.root)
# bt.pre_order_traversal_iterative()
# bt.in_order_traversal_iterative()
# bt.post_order_traversal_iterative_one_stack()
# bt.post_order_traversal_iterative_two_stacks()
bt.delete_node(8)
bt.level_order_traversal_level_by_level()
```
